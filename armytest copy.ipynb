```python
import pdfplumber
import pandas as pd
```

```python
pd.set_option('display.max_colwidth', None)
pd.set_option('expand_frame_repr', False)
```

```python
def get_table(name, pages):
    with pdfplumber.open(name) as pdf:
        tables = []
        for i in pages:
            page = pdf.pages[i]
            t = page.extract_tables()
            if (len(t) == 0):
                continue
            table = t[0]
            tables.append(table)
    return tables
```

```python
missiles = get_table("pdf/m.pdf", range(27,30))
```

```python
print(pd.DataFrame(missiles[0]))
print(pd.DataFrame(missiles[1]))
```
